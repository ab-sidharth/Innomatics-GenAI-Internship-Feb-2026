1.  Employee Performance Bonus Eligibility

```python
ems = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}  # example input dictionary
# for user input 
# n = int(input("Enter number of employees: "))
#  ems = {}
#  for i in range(n):
#      name = input("Enter employee name: ")
#      score = int(input("Enter performance score: "))
#      ems[name] = score)

h_score = max(ems.values())  # finding highest score
top_perf = []
for n in ems:
    if ems[n] == h_score:
        top_perf.append(n)
print("Top Performers Eligible for Bonus:", ", ".join(top_perf), "(Score:", h_score, ")")
```

2.  Search Query Keyword Analysis

```python
q = "Buy mobile phone buy phone online"  # example input
#q = input("Enter search query: ") #for user input
q = q.lower()
wrds = q.split()
fq = {}
# frequency of each keyword
for wrd in wrds:
    if wrd in fq:
        fq[wrd] += 1
    else:
        fq[wrd] = 1
r = {}
for wrd in fq:
    if fq[wrd] > 1:
        r[wrd] = fq[wrd]
print(r)
```

4. Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
] #example input
# for user input
# n = int(input("Enter number of emails: "))
# emails = []
# for i in range(n):
#     email = input("Enter email: ")
#     emails.append(email)
# domain_count = {}
domain_ct = {}
for e in emails:
    dom = e.split("@")[1]
    if dom in domain_ct:
        domain_ct[dom] += 1
    else:
        domain_ct[dom] = 1
ttl = len(emails)
# Calculating the percentage usage of each domain.
for dom in domain_ct:
    per = (domain_ct[dom] / ttl) * 100
    print(dom + ":", int(per), "%")
```

5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]  # example input
#sales = list(map(int, input("Enter daily sales separated by space: ").split())) # for user input
avg = sum(sales) / len(sales) #average daily sales
lt = avg+avg * 0.3  # 30% above average
for i in range(len(sales)):
    if sales[i] >= lt:
        print("Day", i+1, ":", sales[i]) # day number and sale value
```

6. Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# user_ids = input("Enter user IDs separated by space: ").split() # for user input
dict = {}
# for duplicate identification
for uid in user_ids:
    if uid in dict:
        dict[uid] += 1
    else:
        dict[uid] = 1
for uid in dict:
    if dict[uid] > 1:
        print(uid, "→", dict[uid], "times")
```
