1.  Social Media – Post Engagement Analyzer

```python
def alyz_pst(lkes_lst):
    ttl = 0
    for lke in lkes_lst:
        ttl += lke
    print("Total Likes:", ttl)
    if ttl >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

lkes = [200, 300, 150, 400] # example input list
# user input # likes=list(map(int,input("Enter list of likes").split(","))
alyz_pst(lkes)
```

3. Agriculture – Rainfall Adequacy Checker

```python
def rnfal_chkr(rnfal_lst, req_lvl):
    ttl = 0
    for rn in rnfal_lst:
        ttl += rn
    avg = ttl / len(rnfal_lst) # average rainfall level
    print("Average Rainfall:", avg)
    if avg >= req_lvl:# rain evaluation
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
# Taking user input
rnfal_lst = list(map(int, input("Enter rainfall values separated by space: ").split()))
req_lvl = int(input("Enter required rainfall level: "))
rnfal_chkr(rnfal_lst, req_lvl)
```

4. Social Media – Duplicate Account Detection

```python
def chk_dupli(un):
    uq = set(un)
    if len(uq) < len(un):
        print("Duplicate Accounts Found")
    else:
        print(" Accounts Are Unique ")

usrs = ["user1", "user2", "user1", "user3"] # example list
chk_dupli(usrs)
```

7. System Monitoring – Application Health Checker

```python
def chk_sys_h(er_cnt):
    if er_cnt == 0:
        sts = "Healthy"
    elif er_cnt <= 5:
        sts = "Minor Issues"
    else:
        sts = "Critical Issues"
    return sts

ers = int(input("Enter error count: "))
print("Error Count:", ers)
print("System Status:", chk_sys_h(ers))
```

9. E-Learning – Student Attendance Eligibility System

```python
def att_check(att):
    pr = 0
    for d in att:
        if d == "P":
            pr += 1
    pg = (pr / len(att)) * 100
    print("Attendance Percentage:", pg)
    if pg >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

att_check(["P", "P", "A", "P", "P"])
```

10. Smart Electricity Bill Analyzer

```python
def calc_bill(u):
    # Slab calculation
    if u <= 100:
        b = u * 3
    elif u <= 200:
        b = (100 * 3) + ((u - 100) * 5)
    else:
        b = (100 * 3) + (100 * 5) + ((u - 200) * 7)

    # Usage classification
    if b < 500:
        uge = "Low Usage"
    elif b <= 1500:
        uge = "Moderate Usage"
    else:
        uge = "High Usage"
    return b, uge

# Example call
n=int(input("Enter electricity units:"))
ttl_b, uge_lvl = calc_bill(n)

print("Total Bill:", ttl_b)
print("Usage Level:", uge_lvl)
```
