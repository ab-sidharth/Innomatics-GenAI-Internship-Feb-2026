1. Smart Light Controller

```python
class Sm_Lt:
    def __init__(self, nm):
        self.nm = nm
        self.stts = "OFF"
    def trn_on(self):
        self.stts = "ON"
    def trn_off(self):
        self.stts = "OFF"

# Example Input 
# light = SmartLight("Bedroom Light")
# light.turn_on()

# User Input
lt_n = input("Enter Light Name: ")
acn = input("Enter Action (ON/OFF): ")
lt = Sm_Lt(lt_n)
if acn == "ON":
    lt.trn_on()
else:
    lt.trn_off()
print(lt.nm, "is", lt.stts)
```

2. Mobile Contact Record

```python
class Emp:
    def __init__(self, nm, emp_id, dep):
        self.nm = nm
        self.emp_id = emp_id
        self.dep = dep
    def disp_id(self):
        print("Employee ID Card")
        print("Name:", self.nm)
        print("ID:", self.emp_id)
        print("Department:", self.dep)
        
# Example Input
# nm = Rahul
# emp_id = EMP102
# dep = IT

# User Input
nm = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
dep = input("Enter Department: ")
emp = Emp(nm, emp_id, dep)
emp.disp_id()
```

3. Product Price Tag Generator

```python
class Cont:
    def __init__(self, nm, phn):
        self.nm = nm
        self.phn = phn
    def sv_cont(self):
        print("Contact Saved")
        print("Name:", self.nm)
        print("Phone:", self.phn)

# Example Input
# nm = Anita
# phn = 9876543210

# User Input
nm = input("Enter Contact Name: ")
phn = input("Enter Phone Number: ")
cont = Cont(nm, phn)
cont.sv_cont()
```

4. Movie Rating Display System

```python
class Prod:
    def __init__(self, nm, prc):
        self.nm = nm
        self.prc = prc
    def disp_prc(self):
        print("Product:", self.nm)
        print("Price: ₹" + str(self.prc))

# Example Input
# nm = Headphones
# prc = int(2499)

# User Input
nm = input("Enter Product Name: ")
prc = int(input("Enter Price: "))
prod = Prod(nm, prc)
prod.disp_prc()
```

5. Movie Rating Display System

```python
class Mv:
    def __init__(self, nm, ratn):
        self.nm = nm
        self.ratn = ratn
    def disp_mov_ratn(self):
        print("Movie:", self.nm)
        print("Rating:", self.ratn, "/ 5")

# Example Input
# nm = Inception
# ratn = float(4.8)

# User Input
nm = input("Enter Movie Name: ")
ratn = float(input("Enter Rating: "))
mov = Mv(nm, ratn)
mov.disp_mov_ratn()
```

6. Delivery Address Manager

```python
class Del:
    def __init__(self, cust, adr):
        self.cust = cust
        self.adr = adr
    def disp_del(self):
        print("Delivery Details")
        print("Customer:", self.cust)
        print("Address:", self.adr)

# Example Input
# cust = Suman
# adr = Hyderabad

# User Input
cust = input("Enter Customer Name: ")
adr = input("Enter Address: ")
delv = Del(cust, adr)
delv.disp_del()
```
