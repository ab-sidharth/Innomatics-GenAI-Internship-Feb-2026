1. Mobile Recharge Validation System

```python
def validate_r():
    vp = [199, 299, 399, 599] # valid plans

    while True:
        amt = int(input("Enter recharge amount: "))
        if amt < 50:
            print("Invalid amount. Minimum recharge is ₹50.")
        elif amt not in vp:
            print("Invalid plan. Available plans:", vp)
        else:
            print("Recharge Successful!")
            break # checking for a valid recharge

        re = input("Do you want to retry? (yes/no): ").lower() # retry option
        if re != "yes":
            print("Recharge cancelled.")
            break
validate_r()
```

3. Student Result Processing System

```python
def result_status(marks):
    ttl = 0
    for m in marks:
        ttl += m
    avg = ttl / len(marks)  
    print("Average:", avg)
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"
marks = [60, 55, 70, 40] # example input
# marks = list(map(int, input().split(","))) # for user input 
print("Result:", result_status(marks))
```

4. Cab Fare Estimator with Retry Option

```python
def calculate_fare():
    while True:
        km = float(input("Enter distance in km: "))
        p = input("Peak hour? (yes/no): ").lower()
        f = 50 + (km * 12) # ususal fare
        if p == "yes":
            f += f * 0.25 # peak time fare if any
        print("Total Fare: ₹", f)
        
        r = input("Calculate again? (yes/no): ").lower() # retry option
        if r != "yes":
            break

calculate_fare()
```

5. Employee Attendance Eligibility Checker

```python
def check_elig(att):
    pr = 0
    for d in att:
        if d == "P":
            pr += 1
    per = (pr / len(att)) * 100
    print("Attendance %:", per)
    if per >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
att = ["P", "P", "A", "P", "P", "A", "P", "P"] # example input
# a = input("Enter attendance separated by commas: ").split(",") # for user input
print(check_elig(att))
```

6. Password Strength Checker

```python
def check_password(pas):
    dig = False
    spl = False
    if len(pas) < 8:
        return "Weak Password"
    for c in pas:
        if c.isdigit():
            dig = True
        if c in "@#$":
            spl = True
    if dig and spl:
        return "Strong Password"
    else:
        return "Weak Password"
pwd = input("Enter password: ") # example input pwd ="password@1"
print(check_password(pwd))
```
