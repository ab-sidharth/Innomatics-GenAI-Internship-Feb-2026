1. User Login Check

```python
# existing credentials
valid_username = "admin"
valid_password = "1234"
# Test values
input_user = "admin" # or input("Enter username")
input_pass = "1234"  # or input("Enter password) for user input   
# Compare both username and password
if input_user == valid_username and input_pass == valid_password:
    print("Login Successful")
else:
    print("Invalid Credentials") 
```

2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
pas_count = 0
fail_count = 0
for s in marks:
    if s >= 50:
        pas_count += 1
    else:
        fail_count += 1
print("Pass count:", pas_count)
print("Fail count:", fail_count)
```

3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
list = []
for i in  names:
    # Remove extra spaces and convert to lowercase
    list.append(i.strip().lower())
print("After changes :", list)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for string in messages:
    length = len(string)
    print(f"Message: {string} | Length: {length}")
    if length > 10:
        print("Long message detected")
```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count_total = 0
for ntre in logs:
    if ntre == "ERROR":
        error_count_total += 1
print("Number of ERROR logs:", error_count_total)
```
