1. Unique Words in a Sentence

```python
sentence = "python is easy and python is powerful"  # example input, or you can use input()
words = sentence.split()  # splitting the sentence into words
unq_words = []
for word in words:
    if word not in unq_words:
        unq_words.append(word)
# Using a for loop to remove duplicates and keep the first unique occurrence
# We can also use the set() function to remove duplicates instead of a for loop.
print("Unique words count:", len(unq_words))
print("Unique words:", unq_words)
```

2. Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}  # example employee list in dictonary format  
highst_name = ""
highst_salary = 0  # initial values to start comparison
for name in employees:
    if employees[name] > highst_salary:
        highst_salary = employees[name]
        highst_name = name
# Determining the highest salary and the employee name
print("Highest Salary:", highst_name, "-", highst_salary)
```

3. Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]  # initializing list of input values or use list(map(int, input().split())) for input
mx_v= numbers[0]
mn_v = numbers[0]  # initializing max and min values to the first element
for n in numbers:
    if n > mx_v:
        mx_v = n
    if n < mn_v:
        mn_v = n
# Using a loop to compare elements and determine the maximum and minimum values
print("List:", numbers)
print("Max:", mx_v)
print("Min:", mn_v)
```

4. Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]  # example input: list of product prices or use list(map(int, input().split())) for input
ct = 0
for price in prices:
    if price > 1000:
        ct += 1  # using count to track how many prices are greater than 1000
print("Products above 1000:", ct)
```

5. Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]  # example input list containing "P" (Present) and "A" (Absent) or use list(map(str, input().split())) for input
present = 0
for day in attendance:
    if day == "P":
        present += 1
# Calculating the number of present days
percentage = (present / len(attendance)) * 100  # attendance percentage
print("Attendance Percentage:", percentage)
```

7. Count Character Frequency

```python
txt = "python"  # example input string, or else use input()
frq = {}  # dictionary to store character frequencies
for ch in txt:
    if ch in frq:
        frq[ch] += 1
    else:
        frq[ch] = 1
# Using a for loop to determine how many times each character appears
print(frq)
```

10. Calculate Average Marks

```python
mrks = [70, 80, 65, 75]  # example input list of marks
ttl = sum(mrks)
avg = ttl / len(mrks)
# Calculating the sum and average of the list
print("Average Marks:", avg)
```
