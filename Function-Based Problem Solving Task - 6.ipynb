1. Smart Parking Lot Management System

```python
def prk_sys(cap,lgs):
    prk = 0
    pk = 0
    for lg in lgs:
        if lg == "IN":
            prk += 1
        elif lg == "OUT":
            prk -= 1
        if prk > pk:
            pk = prk

    print("Currently Parked Vehicles:", prk)
    if prk > cap:
        print("Parking Status: Full - Over Capacity Alert")
    else:
        print("Parking Status: Available")

# Example Input:
# cap = 50
# lg = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
# User Input:
cap = int(input("Enter Parking Capacity: "))
lgs = input("Enter Vehicle Logs separated by space (IN/OUT): ").split()
prk_sys(cap,lgs)
```

2. Online Food Delivery Time Estimator

```python
def delivery_time(dis, trf, we):
    eta = dis * 5
    if trf == "High":
        eta += 10
    elif trf == "Medium":
        eta += 5
    if we == "Rainy":
        eta += 5

    print("Estimated Delivery Time:", eta, "minutes")
#Example Input:
# dis = 8
# trf = High
# we = Rainy
# User Input:
dis = int(input("Enter Distance (km): "))
trf = input("Enter Traffic Level (Low/Medium/High): ")
we = input("Enter Weather (Clear/Rainy): ")
delivery_time(dis, trf, we)
```

3. Movie Theatre Seat Occupancy Analyzer

```python
def tt_occ(ttl_s,bk_cnt):
    occ = (len(bk_cnt) / ttl_s) * 100
    print("Occupancy:", int(occ), "%")

    if occ == 100:
        print("Show Status: Housefull")
    elif occ >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

# Example Input:
# tt_s = 200
# bk_cnt = [1]*150   
ttl_s = int(input("Enter Total Seats: "))
bk_cnt = [1]*int(input("Enter Number of Booked Seats: "))
tt_occ(ttl_s,bk_cnt)
```

4. Cloud Server Load Classification System

```python
def s_ld(c_r):
    ttl = 0
    for vl in c_r:
        ttl += vl
    ag = ttl / len(c_r)
    print("Average CPU Load:", int(ag), "%")

    if ag < 50:
        print("Server Status: Normal")
    elif ag <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
        
c_r = list(map(int, input("Enter CPU Readings separated by space: ").split()))# user input
# c_r = [45, 60, 70, 85, 90] # example input
s_ld(c_r)
```

5. Smart Classroom Resource Usage Monitor

6. Online Event Registration Capacity Controller

```python
def ev_reg(cap,re):
    if re >= cap:
        cnfm = cap
        wl = re - cap
        st = "Closed"
    else:
        cnfm = re
        wl = 0
        st = "Open"
    print("Confirmed Registrations:", cnfm)
    print("Waitlisted Users:", wl)
    print("Registration Status:", st)

# Example Input:
# Event Capacity: 100
# Registrations: 105
c = int(input("Enter Event Capacity: "))
r = int(input("Enter Total Registrations: "))
ev_reg(c,r)
```
